```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize
stemmer = PorterStemmer()
sentences = sent_tokenize(paragraph)
for sentence in sentences: 
    print(sentence.strip())
```

```python
# Apply Stopwords and filter out stopwords

for i in range(len(sentences)):
    # convert the sentences to words
    words = nltk.word_tokenize(sentences[i])
    # go through each word, check if the word is a stopword, stem the word
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    #replace the words with their stems
    sentences[i] = ' '.join(words)

sentences
```

```python
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')

for i in range(len(sentences)):
    # convert the sentences to words
    words = nltk.word_tokenize(sentences[i])
    # go through each word, check if the word is a stopword, stem the word
    words = [snowball_stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    #replace the words with their stems
    sentences[i] = ' '.join(words)

sentences
```
